```python
import json
from bertopic import BERTopic
```

```python
def load_docs(path: str = './corpus.jsonl'):
  with open(path) as f:
    return [json.loads(jline)["text"] for jline in f.read().splitlines()]
```

```python
docs = load_docs('/content/corpus.jsonl')
```

```python
docs_topic_model = BERTopic(calculate_probabilities=True, verbose=True)
docs_topics, docs_probs = docs_topic_model.fit_transform(docs)
```
